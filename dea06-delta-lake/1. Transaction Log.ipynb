# Understanding delta lake transaction log

## 0. Create a new schema under the demo catalog for this section of the course (delta_lake)

```python
%sql
-- create the demo catalog
CREATE CATALOG IF NOT EXISTS demo
MANAGED LOCATION 'abfss://demo@maideacourseextdl.dfs.core.windows.net/'
```

```python
%sql
-- create schema under demo catalog
CREATE SCHEMA IF NOT EXISTS demo.delta_lake
MANAGED LOCATION 'abfss://demo@maideacourseextdl.dfs.core.windows.net/delta_lake'
```

## 1. Create a delta lake table

```python
%sql
CREATE TABLE IF NOT EXISTS demo.delta_lake.companies
(
  company_name STRING,
  founded_date DATE,
  country STRING
);
```

```python
%sql
DESCRIBE EXTENDED demo.delta_lake.companies;
```

## 2. Insert some data

```python
%sql 
INSERT INTO demo.delta_lake.companies
VALUES ("Apple", "1976-04-01", "USA");
```

```python
%sql
SELECT *
FROM demo.delta_lake.companies;
```
