# Remove unused files using VACUUM
## VACUUM Command

## 1. Check table history

```python
%sql
desc history demo.delta_lake.optimize_stock_prices;
```

## 2. Run VACUUM

```python
%sql
VACUUM demo.delta_lake.optimize_stock_prices
```

```python
%sql
desc history demo.delta_lake.optimize_stock_prices;
```

```python
%sql
-- warning: DO NOT DO THIS IN PRODUCTION
SET spark.databricks.delta.retentionDurationCheck.enabled = false;

VACUUM demo.delta_lake.optimize_stock_prices
RETAIN 0 HOURS;
```

```python
%sql
desc history demo.delta_lake.optimize_stock_prices;
```

```python
%sql
-- latest data still available
select * from demo.delta_lake.optimize_stock_prices;
```

```python
%sql
-- but previous versions (history) will no longer be available
select * from demo.delta_lake.optimize_stock_prices
version as of 3;
```
