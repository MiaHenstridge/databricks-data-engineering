# History and Time Travel
1. Query Delta Lake table history
2. Query previous versions of the data
3. Query data from a specific time
4. Restore data to a specific version

## 1. Query Delta lake table history

```python
%sql
DESCRIBE HISTORY demo.delta_lake.companies;
```

## 2. Query previous versions of data

```python
%sql 
SELECT * 
FROM demo.delta_lake.companies
VERSION AS OF 1;
```

## 3. Query data from a specific time

```python
%sql
SELECT *
FROM demo.delta_lake.companies
TIMESTAMP AS OF '2026-01-30T13:55:23.000+00:00';
```

## 4. Restore data to a specific version

```python
%sql
RESTORE TABLE demo.delta_lake.companies
VERSION AS OF 1;
```
