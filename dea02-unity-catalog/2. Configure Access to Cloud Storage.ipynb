# Configure Access to Cloud Storage via Unity Catalog

## Access Cloud Storage

```python
%fs ls 'abfss://demo@maideacourseextdl.dfs.core.windows.net/'
```

## Create External Location

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dea_course_ext_dl_demo
  URL 'abfss://demo@maideacourseextdl.dfs.core.windows.net/'
  WITH (STORAGE CREDENTIAL dea-course-ext-access-credential)
  COMMENT "External Location for demo purposes"
```
