# Configure access to Azure SQL database via Lakehouse Federation
1. Create connection
2. Create foreign catalog

Documentation: https://docs.databricks.com/aws/en/query-federation/database-federation

```python
%sql
-- create connection
CREATE CONNECTION sql_gizmobox_db_conn_sql TYPE sqlserver
OPTIONS (  
  host 'mai-gizmobox-srv.database.windows.net',
  port '1433',
  user 'gizmoboxadm',
  password 'Thespacebard@22' -- recommended: use Databricks secrets
);
```

```python
# %sql
# DROP CATALOG IF EXISTS sql_gizmobox_db_catalog_sql;
# DROP CONNECTION sql_gizmobox_db_conn_sql;
```

```python
%sql
CREATE FOREIGN CATALOG IF NOT EXISTS sql_gizmobox_db_catalog_sql 
USING CONNECTION sql_gizmobox_db_conn_sql
OPTIONS (database 'gizmobox-db');
```

```python
%sql
SHOW CATALOGS;

USE CATALOG sql_gizmobox_db_catalog_sql;
USE SCHEMA dbo;
SHOW TABLES;

SELECT * FROM refunds;
```

```python
%sql 
-- join the refunds table with existing tables in the catalog
SELECT p.*, r.*
FROM sql_gizmobox_db_catalog_sql.dbo.refunds r
JOIN gizmobox.silver.payments p on r.payment_id = p.payment_id
```
