## Demo 1
Demonstrate
1. Breakpoints
2. Step through code line by line
3. Variable Explorer
4. Debug Console

```python
# Calculate the final price of an item and print that value
item_price = 120.00
tax_rate = 20       # percentage
discount = 10.00

# 1. Calculate the tax to be applied
tax = item_price * tax_rate / 100

# 2. Apply tax to item price
item_price_with_tax = item_price + tax

# 3. Apply discount to item price
final_price = item_price_with_tax - discount

# print final price
print(f"Final price: {final_price}")
```

## Demo 2
Demonstrate Step in and Step out

```python
def calculate_final_price(item_price, tax_rate, discount):
    tax = item_price * tax_rate / 100
    item_price_with_tax = item_price + tax
    final_price = item_price_with_tax - discount
    return final_price
```

```python
# main program
item_price = 170.00
tax_rate = 20
discount = 10.00

final_price = calculate_final_price(item_price, tax_rate, discount)
print(f"Final price: {final_price}")
```
