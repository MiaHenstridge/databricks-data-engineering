# Data profiling in Databricks
1. Profilng data using UI
2. Profile data using dbutils package
3. Profile data manually
  * COUNT
  * COUNT_IF
  * MIN
  * MAX
  * WHERE clause

## 1. Profile data using UI

```python
%sql
SELECT *
FROM gizmobox.bronze.v_customers;
```

## 2. Profile data using DBUTILS package (dbutils.data.summarize method)

```python
df = spark.table('gizmobox.bronze.v_customers')

dbutils.data.summarize(df)
```

## 3. Manual profiling

### 3.1. COUNT function

```python
%sql
SELECT 
  COUNT(*), 
  COUNT(customer_id), 
  COUNT(email), 
  COUNT(telephone)
FROM gizmobox.bronze.v_customers;
```

### 3.2. COUNT_IF function

```python
%sql
SELECT
  COUNT(*),
  count_if(customer_id is NULL),
  count_if(email is NULL),
  count_if(telephone is NULL)
FROM gizmobox.bronze.v_customers;
```

### 3.3. WHERE clause

```python
%sql
SELECT COUNT(*)
FROM gizmobox.bronze.v_customers
WHERE customer_id IS NULL;  
```

### 3.4. DISTINCT keyword

```python
%sql
SELECT COUNT(*), COUNT(DISTINCT customer_id)
FROM gizmobox.bronze.v_customers
WHERE customer_id IS NOT NULL;
```
