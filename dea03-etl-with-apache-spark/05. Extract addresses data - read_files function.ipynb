# Extract data from the address files
1. Demonstrate limitations of CSV file format in SELECT statement
2. Use read_files function to overcome the limitation
3. Create Addresses view in the Bronze schema

## 1. Demonstrate limitations of csv file format in SELECT statement

```sql
SELECT *
FROM csv.`/Volumes/gizmobox/landing/operational_data/addresses/`;
```

## 2. Read_files function
Documentation: https://learn.microsoft.com/en-us/azure/databricks/sql/language-manual/functions/read_files

```sql
SELECT * 
FROM read_files(
  '/Volumes/gizmobox/landing/operational_data/addresses/', 
  format =>'csv',
  header => true,
  delimiter => '\t');
```

## 3. Create Addresses view in the Bronze layer

```sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_addresses 
AS 
SELECT *
FROM read_files(
  '/Volumes/gizmobox/landing/operational_data/addresses/', 
  format =>'csv',
  header => true,
  delimiter => '\t');
```

```sql
SELECT *
FROM gizmobox.bronze.v_addresses;
```
