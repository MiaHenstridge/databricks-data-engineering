# Query orders data as JSON strings
1. Extract top level column values
2. Extract array elements
3. Extract nested column values
4. Cast column values as a specific data type

```python
%sql
SELECT *
FROM gizmobox.bronze.v_orders;
```

## 1. Extract top level info

```python
%sql
select
  value:order_id as order_id,
  value:customer_id as customer_id,
  value:order_date as order_date,
  value:transaction_timestamp as transaction_timestamp,
  value:total_amount as total_amount
from gizmobox.bronze.v_orders;
```

## 2. Extract array elements

```python
%sql
select
  value:items[0] as items,
  value
from gizmobox.bronze.v_orders;
```

## 3. Extract nested column values

```python
%sql
select
  value:items[0].item_id::int as item_1_item_id,
  value
from gizmobox.bronze.v_orders;
```
