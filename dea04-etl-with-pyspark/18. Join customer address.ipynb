# Join customer and address
Join customer data with address data to create a customer_address table which contains the address of each customer on the same record

```python
df_customers = spark.table('gizmobox.silver.py_customers')
```

```python
df_addresses = spark.table('gizmobox.silver.py_addresses')
```

```python
df_customer_w_address = df_customers.join(df_addresses, df_customers.customer_id == df_addresses.customer_id)\
                                      .select(
                                        df_customers.customer_id,
                                        df_customers.email,
                                        df_customers.date_of_birth,
                                        df_customers.member_since,
                                        df_customers.telephone,
                                        df_addresses.shipping_address_line_1,
                                        df_addresses.shipping_city,
                                        df_addresses.shipping_state,
                                        df_addresses.shipping_postcode,
                                        df_addresses.billing_address_line_1,
                                        df_addresses.billing_city,
                                        df_addresses.billing_state,
                                        df_addresses.billing_postcode
                                      )

display(df_customer_w_address.limit(10))
```

```python
df_customer_w_address.writeTo('gizmobox.gold.py_customer_address').createOrReplace()
```
