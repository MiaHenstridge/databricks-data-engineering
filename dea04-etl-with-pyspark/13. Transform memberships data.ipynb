# Transform Memberships data
1. Extract customer_id from the file path
2. Write transformed data to the Silver schema

## 1. Extract customer_id from the file path

```python
import pyspark.sql.functions as F
import pyspark.sql.types as T

df = spark.read.table("gizmobox.bronze.py_memberships")\
        .select(
          F.regexp_extract('path', '.*/([0-9]+)\\.png$', 1).alias('customer_id'),
          F.col('content').alias('membership_card')
        )

display(df)
```

## 2. Write transformed data to the Silver schema

```python
df.writeTo("gizmobox.silver.py_memberships").createOrReplace()
```
