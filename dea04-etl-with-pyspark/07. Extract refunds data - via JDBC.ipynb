# Extract data from the Refunds SQL Table
1. Read refunds data via JDBC
2. Create refunds table in bronze layer

NOTE: https://www.udemy.com/course/databricks-certified-data-engineer-associate-ultimate-prep/learn/lecture/54097825#content

## 1. Read refunds data via JDBC

```python
df = spark.read.format('jdbc')\
    .option('url', 'jdbc:sqlserver://mai-gizmobox-srv.database.windows.net:1433;database=gizmobox-db')\
    .option('dbtable', 'refunds')\
    .option('user', 'gizmoboxadm')\
    .option('password', 'Thespacebard@22')\
    .load()

display(df)
```

## 2. Write the refunds data to the bronze schema

```python
df.writeTo('gizmobox.bronze.py_refunds').createOrReplace()
```
