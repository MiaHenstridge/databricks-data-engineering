## 1. Create external location
* External location name: dea_course_ext_dl_circuitbox
* ADLS path: abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/
* Storage credential: dea-course-ext-access-credential

```python
%sql
create external location if not exists dea_course_ext_dl_circuitbox
url 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/'
with (storage credential `dea-course-ext-access-credential`)
comment 'External location for the circuitbox data lakehouse';
```

```python
%fs ls 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/'
```

## 2. Create catalog
* Catalog name: circuitbox
* Managed location: 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/'

```python
%sql
create catalog if not exists circuitbox
managed location 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/'
comment 'Catalog for the circuitbox data lakehouse';
```

```python
%sql
show catalogs;
```

## 3. Create 
1. Schema name: landing
* Managed location: 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/landing'

2. Schema name: lakehouse
* Managed location: 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/lakehouse'

```python
%sql
use catalog circuitbox;

create schema if not exists landing
managed location 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/landing';

create schema if not exists lakehouse
managed location 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/lakehouse';
```

```python
%sql
show schemas;
```

## 4. Create Volume
Volume name: operational_data

ADLS path: 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/landing/operational_data'

```python
%sql
use catalog circuitbox;
use schema landing;

create external volume if not exists operational_data
location 'abfss://circuitbox@maideacourseextdl.dfs.core.windows.net/landing/operational_data';
```
